```python
import math
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.sinyi.com.tw/tradeinfo/list/Taipei-city/106-zip/6month-dealtime/datatime-desc/1"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
def getDataCount(soup):
    container = soup.find("div", class_="trade-list-map-title")
    first_span = container.select("span")[0]
    second_span = first_span.select_one("span")
    return int(second_span.getText())

data_count = getDataCount(soup)
print(data_count)
```

```python
def getRowData(container):
    child_container = container.select("div")
    data = {
        "soldDate": child_container[0].getText(),
        #"address": child_container[1].getText(),
        #"type": child_container[2].getText(),
        #"totalPrice": child_container[3].getText(),
        "uniPrice": child_container[4].getText(),
        #"areaBuilding": child_container[5].getText(),
        #"areaLand": child_container[6].getText(),
    }
    return data
    #for idx, content in enumerate(child_container):
    #    print(f"id: {idx}, content: {content.getText()}")

def getTableData(soup):
    container = soup.find_all("div", class_="trade-obj-card-web")
    tableData = []
    for idx, content in enumerate(container):
        tableData.append(getRowData(content))
    return tableData
```

```python
url = f"https://www.sinyi.com.tw/tradeinfo/list/Taipei-city/106-zip/6month-dealtime/datatime-desc/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
totalPage = math.ceil(getDataCount(soup)/5)
print(totalPage)
data = []
i = 1
while (i <= 5):
    print(i)
    url = f"https://www.sinyi.com.tw/tradeinfo/list/Taipei-city/106-zip/6month-dealtime/datatime-desc/{i}"
    response = requests.get(url)
    print(response)
    soup = BeautifulSoup(response.text, "html.parser")
    data.extend(getTableData(soup))
    #print(getTableData(soup))
    i = i + 1
    
print(data)
```
