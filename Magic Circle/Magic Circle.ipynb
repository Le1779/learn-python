```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def drawCircle(r, lineWidth, x = 0, y = 0):
    t = np.linspace(0, 2* np.pi, 101)
    plt.plot(r*np.cos(t) + x, r*np.sin(t) + y, 'w', linewidth=lineWidth)
    
def drawText(x, y, text, size, rotation=0):
    ax.text(x, y, text, color='white', va='center', ha='center', fontsize=size, rotation=rotation)

def drawLine(x1, y1, x2, y2, lineWidth):
    plt.plot([x1, x2], [y1, y2], 'w', linewidth=lineWidth)

def initCanvas(w):
    plt.figure(figsize=(w,w))
    
def setBlackBackground():
    ax.set_facecolor('black')

def drawOutline(w):
    lineWidth = w*0.3
    
    r = w
    drawCircle(r, lineWidth)
    
    r = w*0.95
    drawCircle(r, lineWidth)
    
    r = w*0.8
    drawCircle(r, lineWidth)
    
    text = "Φ†♂✴☥☪"
    for i in range(len(text)):
        s = np.pi*2*i/len(text) + np.pi*0.5
        r = w*0.88
        x0,y0 = r*np.cos(s), r*np.sin(s)
        drawText(x0, y0, text[i], w*3)
        
        subText = "made by kevin le 2020/10/16"
        for si in range(len(subText)):
            index = (len(subText)-si)/len(subText)
            ss = np.pi/5*index + s + np.pi*0.05
            sr = w*0.83
            sx0,sy0 = sr*np.cos(ss), sr*np.sin(ss)
            drawText(sx0, sy0, subText[si], w*0.6, ss*180/np.pi-90)
    
def drawInline(w):
    lineWidth = w*0.15
    
    r = w*0.65
    drawCircle(r, lineWidth)
    
    r = w*0.25
    drawCircle(r, lineWidth)
    
    r = w*0.2
    drawCircle(r, lineWidth)
    
    for i in range(6):
        s = np.pi*2*i/6 + np.pi*0.5
        x,y = r*np.cos(s), r*np.sin(s)
        drawLine(x, y, 0, 0, lineWidth)
    
    for i in range(4):
        s = np.pi*2*i/4
        subText = "made by kevin le"
        for si in range(len(subText)):
            index = (len(subText)-si)/len(subText)
            ss = np.pi/2*index + s
            sr = w*0.22
            sx0,sy0 = sr*np.cos(ss), sr*np.sin(ss)
            drawText(sx0, sy0, subText[si], w*0.5, ss*180/np.pi-90)
            
    for i in range(3):
        s = np.pi*2*i/3 - np.pi*0.1
        subText = "made by kevin le"
        for si in range(len(subText)):
            index = (len(subText)-si)/len(subText)
            ss = np.pi/2*index + s
            sr = w*0.4
            sx0,sy0 = sr*np.cos(ss), sr*np.sin(ss)
            drawText(sx0, sy0, subText[si], w*0.5, ss*180/np.pi-90)
            
def drawHex(w):
    t = np.linspace(np.pi*0.5, 2.5* np.pi, 7)
    r = w*0.8
    plt.plot(r*np.cos(t), r*np.sin(t), 'w', linewidth=w*0.15)
    
def drawInnerCircle(w):
    lineWidth = w*0.15
    pr = w*0.45
    r = w*0.1
    
    text = "ΣξΘ"
    for i in range(3):
        s = np.pi*2*i/3 + np.pi*0.5
        x,y = pr*np.cos(s), pr*np.sin(s)
        drawCircle(r, lineWidth, x, y)
        drawText(x, y, text[i], w*3.5, s*180/np.pi-90)
        
def draw3PartOfCircle(w):
    lineWidth = w*0.3
    pr = w*0.8
    r = w*0.33
    
    slw = w*0.15
    sr = w*0.25
    
    for i in range(3):
        s = np.pi*2*i/3 + np.pi*1/6
        x,y = pr*np.cos(s), pr*np.sin(s)
        
        angle = 2*np.pi/3
        startAngle = 2*np.pi*0.418
        t = np.linspace(startAngle + angle*i, startAngle + angle + angle*i, 101)
        plt.plot(r*np.cos(t) + x, r*np.sin(t) + y, 'w', linewidth=lineWidth)
        
        plt.plot(sr*np.cos(t) + x, sr*np.sin(t) + y, 'w', linewidth=slw)
        
        subText = "made by kevin le"
        for si in range(len(subText)):
            index = (len(subText)-si)/len(subText)
            ss = np.pi*2*0.3*index + startAngle + angle*i
            tr = w*0.27
            drawText(tr*np.cos(ss) + x, tr*np.sin(ss) + y, subText[si], w*0.7, ss*180/np.pi-90)
            
def drawInnerLine(w):
    lineWidth = w*0.15
    largeLW = w*0.3
    outPosX = []
    outPosY = []
    r = w*0.8
    for i in range(6):
        s = np.pi*2*i/6 + np.pi*0.5
        x,y = r*np.cos(s), r*np.sin(s)
        outPosX.append(x)
        outPosY.append(y)
        
    innerPosX = []
    innerPosY = []
        
    pr = w*0.45
    r = w*0.1
    for i in range(3):
        s = np.pi*2*i/3 + np.pi*0.5
        x,y = pr*np.cos(s), pr*np.sin(s)
        
        posX = []
        posY = []
        for j in range(4):
            js = np.pi*2*j/4 + np.pi*2/3*i
            jx,jy = r*np.cos(js), r*np.sin(js)
            posX.append(jx+x)
            posY.append(jy+y)
        innerPosX.append(posX)
        innerPosY.append(posY)
        
    drawLine(outPosX[0], outPosY[0], innerPosX[0][1], innerPosY[0][1], lineWidth)
    
    drawLine(outPosX[1], outPosY[1], innerPosX[0][2], innerPosY[0][2], largeLW)
    drawLine(outPosX[1], outPosY[1], innerPosX[1][0], innerPosY[1][0], largeLW)
    
    drawLine(outPosX[2], outPosY[2], innerPosX[1][1], innerPosY[1][1], lineWidth)
    
    drawLine(outPosX[3], outPosY[3], innerPosX[1][2], innerPosY[1][2], largeLW)
    drawLine(outPosX[3], outPosY[3], innerPosX[2][0], innerPosY[2][0], largeLW)
    
    drawLine(outPosX[4], outPosY[4], innerPosX[2][1], innerPosY[2][1], lineWidth)
    
    drawLine(outPosX[5], outPosY[5], innerPosX[2][2], innerPosY[2][2], largeLW)
    drawLine(outPosX[5], outPosY[5], innerPosX[0][0], innerPosY[0][0], largeLW)
    
    #close outline line
        
    partInnerPosX = []
    partInnerPosY = []
    r = w*0.55
    for i in range(3):
        s = np.pi*2*i/3 + np.pi*1/6
        x,y = r*np.cos(s), r*np.sin(s)
        partInnerPosX.append(x)
        partInnerPosY.append(y)
        
    drawLine(outPosX[1], outPosY[1], partInnerPosX[1], partInnerPosY[1], lineWidth)
    drawLine(outPosX[3], outPosY[3], partInnerPosX[2], partInnerPosY[2], lineWidth)
    drawLine(outPosX[5], outPosY[5], partInnerPosX[0], partInnerPosY[0], lineWidth)
    
    partOutPosX = []
    partOutPosY = []
    r = w*0.47
    for i in range(3):
        s = np.pi*2*i/3 + np.pi*1/6
        x,y = r*np.cos(s), r*np.sin(s)
        partOutPosX.append(x)
        partOutPosY.append(y)
    
    r = w*0.25
    centerPosX = []
    centerPosY = []
    for i in range(6):
        s = np.pi*2*i/6 + np.pi*0.5
        x,y = r*np.cos(s), r*np.sin(s)
        centerPosX.append(x)
        centerPosY.append(y)
        
    drawLine(centerPosX[0], centerPosY[0], innerPosX[0][3], innerPosY[0][3], lineWidth)
    drawLine(centerPosX[1], centerPosY[1], partOutPosX[1], partOutPosY[1], lineWidth)
    drawLine(centerPosX[2], centerPosY[2], innerPosX[1][3], innerPosY[1][3], lineWidth)
    drawLine(centerPosX[3], centerPosY[3], partOutPosX[2], partOutPosY[2], lineWidth)
    drawLine(centerPosX[4], centerPosY[4], innerPosX[2][3], innerPosY[2][3], lineWidth)
    drawLine(centerPosX[5], centerPosY[5], partOutPosX[0], partOutPosY[0], lineWidth)
```

```python
w = 20
initCanvas(w)
ax = plt.gca()
setBlackBackground()
drawOutline(w)
drawInline(w)
drawInnerCircle(w)
drawHex(w)
draw3PartOfCircle(w)
drawInnerLine(w)


plt.savefig('magic circle.png')
```
